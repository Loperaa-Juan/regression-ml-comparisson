## **NEURAL NETWORKS**

First, import the necessary libraries. In this case, we will need to import tensorflow, keras and numpy.

```python
import tensorflow as tf
import numpy as np
import pandas as pd
from packaging import version

assert version.parse(tf.__version__) >= version.parse("2.8.0")
```

Before training the models, let's charge the dataset. This is the most important step of the process because it is the data that will be used to train the models.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv("/content/drive/MyDrive/Student_Performance.csv")
```

In this case, the dataset contains a categorical variable, the Extracurricular activities of the student. We will map the categorical values to numerical ones, so we can use it in our models. We will map `Yes` to 1 and `No` to 0.

```python
df['Extracurricular Activities'] = df['Extracurricular Activities'].map({'Yes': 1, 'No': 0})
```

To train the model, we need to split the data into features and labels. In this case the features are the columns of the dataset except the last one, and the label is the last column (Performance Index of the student).

```python
X = df.drop("Performance Index", axis=1)
Y = df['Performance Index']
```

Now we will split the data into training and testing sets.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.35, random_state=42)
```

We need set the seed for the random number generator to ensure reproducibility.

```python
tf.random.set_seed(42)
```

We define the models to train, in this case, we will create and train three models:
1. A rectangular model with 3 hidden layers and 10 neurons in each layer.
2. A triangular model with 3 hidden layers and 32, 16, and 8 neurons respectively.
3. A diamond model with 3 hidden layers and 16, 32, and 16 neurons respectively.

## **First Model**

We create a first model with three hidden layers of 10 neurons each, and a linear output layer. (Rectangular architecture)

```python
tf.keras.backend.clear_session()

first_model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation="relu", input_shape=[5, ]),
    tf.keras.layers.Dense(10, activation="relu"),
    tf.keras.layers.Dense(10, activation="relu"),
    tf.keras.layers.Dense(1, activation="linear"),
])
```

Let's see the architecture of the model. we use the `summary()` method to print a summary of the model.

```python
first_model.summary()
```

## **SECOND MODEL**

```python
tf.keras.backend.clear_session()

second_model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation="relu", input_shape=[5, ]),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(8, activation="relu"),
    tf.keras.layers.Dense(1, activation="linear"),
])
```

Like in the previous example, we will use the `summary()` method to print the architecture of the model.

```python
second_model.summary()
```

## **Third Model**

```python
tf.keras.backend.clear_session()

third_model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation="relu", input_shape=[5, ]),
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(1, activation="linear"),
])
```

```python
third_model.summary()
```

Now let's compile and train the models.

```python
first_model.compile(loss="mae", optimizer="adam", metrics=["mae"])
```

```python
second_model.compile(loss="mae", optimizer="adam", metrics=["mae"])
```

```python
third_model.compile(loss="mae", optimizer="adam", metrics=["mae"])
```

```python
first_model.fit(X_train,
                y_train,
                batch_size=32,
                steps_per_epoch=int(np.ceil(len(X_train) / float(32))),
                validation_steps=int(np.ceil(len(X_test) / float(32))),
                epochs=100)
```

```python
second_model.fit(X_train,
                y_train,
                batch_size=32,
                steps_per_epoch=int(np.ceil(len(X_train) / float(32))),
                validation_steps=int(np.ceil(len(X_test) / float(32))),
                epochs=100)
```

```python
# Training the second model (no dropout)
third_model.fit(X_train,
                y_train,
                batch_size=32,
                steps_per_epoch=int(np.ceil(len(X_train) / float(32))),
                validation_steps=int(np.ceil(len(X_test) / float(32))),
                epochs=100)
```

In this case, we will evaluate the performance of the models using the test set. The best model will be used in our web service.

```python
first_model.evaluate(X_test, y_test)
```

```python
second_model.evaluate(X_test, y_test)
```

```python
third_model.evaluate(X_test, y_test)
```

## **METRICS**
The metric we will use to evaluate the performance of the models is the Mean Absolute Error (MAE). This metric measures the average absolute error between the predicted values and the true values. The lower the MAE, the better the model.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
y_pred_first = first_model.predict(X_test)
y_pred_second = second_model.predict(X_test)
y_pred_third = third_model.predict(X_test)
```

First, we will evaluate all the scores of the first model, then we will evaluate the second model and finally the third model. We will compare the results of the three models to see which one performs better.

```python
mse_first_model = mean_squared_error(y_test, y_pred_first)
mae_first_model = mean_absolute_error(y_test, y_pred_first)
r2_first_model = r2_score(y_test, y_pred_first)

print("First Model - MAE:", mae_first_model)
print("First Model - MSE:", mse_first_model)
print("First Model - R2 Score:", r2_first_model)
```

```python
mse_second_model = mean_squared_error(y_test, y_pred_second)
mae_second_model = mean_absolute_error(y_test, y_pred_second)
r2_second_model = r2_score(y_test, y_pred_second)

print("Second Model - MAE:", mae_second_model)
print("Second Model - MSE:", mse_second_model)
print("Second Model - R2 Score:", r2_second_model)
```

```python
mse_third_model = mean_squared_error(y_test, y_pred_third)
mae_third_model = mean_absolute_error(y_test, y_pred_third)
r2_third_model = r2_score(y_test, y_pred_third)

print("Third Model - MAE:", mae_third_model)
print("Third Model - MSE:", mse_third_model)
print("Third Model - R2 Score:", r2_third_model)
```

As we can see, the second model without dropout performed better than the other models. This model has a lower loss and a lower MAE, indicating that it is better at making predictions.
