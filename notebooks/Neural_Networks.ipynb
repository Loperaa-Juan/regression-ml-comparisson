## **NEURAL NETWORKS**

First, import the necessary libraries. In this case, we will need to import tensorflow, keras and numpy.

```python
import tensorflow as tf
import numpy as np
import pandas as pd
from packaging import version

assert version.parse(tf.__version__) >= version.parse("2.8.0")
```

Before training the models, let's charge the dataset. This is the most important step of the process because it is the data that will be used to train the models.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv("/content/drive/MyDrive/Student_Performance.csv")
```

In this case, the dataset contains a categorical variable, the Extra-curricular activities of the student. We will drop this variable from the dataset.

```python
df = df.drop("Extracurricular Activities", axis=1)
```

To train the model, we need to split the data into features and labels. In this case the features are the columns of the dataset except the last one, and the label is the last column (Performance Index of the student).

```python
X = df.drop("Performance Index", axis=1)
Y = df['Performance Index']
```

Now we will split the data into training and testing sets.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=42)
```

We need set the seed for the random number generator to ensure reproducibility.

```python
tf.random.set_seed(42)
```

We define the models to train, in this case, we will create and train four models: two simple models (sequentials) and two models with the same architecture but adding dropout layers.

## **First and Second Model**

We create a first model with three hidden layers of 10 neurons each, and a linear output layer. (Rectangular architecture)

```python
tf.keras.backend.clear_session()

first_model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation="relu", input_shape=[4, ]),
    tf.keras.layers.Dense(10, activation="relu"),
    tf.keras.layers.Dense(10, activation="relu"),
    tf.keras.layers.Dense(1, activation="linear"),
])
```

Let's see the architecture of the model. we use the `summary()` method to print a summary of the model.

```python
first_model.summary()
```

Let's define the second model. In this case, we will use a triangular architecture. Three layers with 32, 16 and 8 neurons respectively.

```python
tf.keras.backend.clear_session()

second_model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation="relu", input_shape=[4, ]),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(8, activation="relu"),
    tf.keras.layers.Dense(1, activation="linear"),
])
```

Like in the previous example, we will use the `summary()` method to print the architecture of the model.

```python
second_model.summary()
```

## **Third and Fourth Model**

Now we will create a third and fourth model with the same architecture as the first and second model, but adding a dropout layer to prevent overfitting.

```python
tf.keras.backend.clear_session()

first_model_dropout = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation="relu", input_shape=[4, ]),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.Dense(10, activation="relu"),
    tf.keras.layers.Dense(10, activation="relu"),
    tf.keras.layers.Dense(1, activation="linear"),
])
```

```python
first_model_dropout.summary()
```

```python
tf.keras.backend.clear_session()

second_model_dropout = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation="relu", input_shape=[4, ]),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(8, activation="relu"),
    tf.keras.layers.Dense(1, activation="linear"),
])
```

```python
second_model_dropout.summary()
```

Now let's compile and train the models.

```python
first_model.compile(loss="mae", optimizer="adam", metrics=["mae"])
first_model_dropout.compile(loss="mae", optimizer="adam", metrics=["mae"])
```

```python
second_model.compile(loss="mae", optimizer="adam", metrics=["mae"])
second_model_dropout.compile(loss="mae", optimizer="adam", metrics=["mae"])
```

```python
# Training the first model (no dropout)

first_model.fit(X_train,
                y_train,
                batch_size=32,
                steps_per_epoch=int(np.ceil(len(X_train) / float(32))),
                validation_steps=int(np.ceil(len(X_test) / float(32))),
                epochs=100)
```

```python
# Training the first model (with dropout)
first_model_dropout.fit(X_train,
                y_train,
                batch_size=32,
                steps_per_epoch=int(np.ceil(len(X_train) / float(32))),
                validation_steps=int(np.ceil(len(X_test) / float(32))),
                epochs=100)
```

```python
# Training the second model (no dropout)
second_model.fit(X_train,
                y_train,
                batch_size=32,
                steps_per_epoch=int(np.ceil(len(X_train) / float(32))),
                validation_steps=int(np.ceil(len(X_test) / float(32))),
                epochs=100)
```

```python
second_model_dropout.fit(X_train,
                y_train,
                batch_size=32,
                steps_per_epoch=int(np.ceil(len(X_train) / float(32))),
                validation_steps=int(np.ceil(len(X_test) / float(32))),
                epochs=100)
```

In this case, we will evaluate the performance of the models using the test set. The best model will be used in our web service.

```python
first_model.evaluate(X_test, y_test)
```

```python
first_model_dropout.evaluate(X_test, y_test)
```

```python
second_model.evaluate(X_test, y_test)
```

```python
second_model_dropout.evaluate(X_test, y_test)
```

As we can see, the second model without dropout performed better than the other models. This model has a lower loss and a lower MAE, indicating that it is better at making predictions.
